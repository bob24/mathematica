```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3, 3, 100)
f = lambda x: x**2
g = lambda x: np.exp(x)

plt.figure(figsize=(8, 4))
plt.plot(x, f(x), label="f(x) = x²")
plt.plot(x, g(x), label="g(x) = eˣ")
plt.title("Examples of Convex Functions")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

# Convex set (circle)
theta = np.linspace(0, 2 * np.pi, 100)
x1, y1 = np.cos(theta), np.sin(theta)
ax[0].fill(x1, y1, alpha=0.5, label='Convex Set')
ax[0].plot([0.5, -0.5], [0.5, -0.5], 'r--', label='Line between two points')
ax[0].set_title("Convex Set")
ax[0].legend()
ax[0].axis('equal')

# Non-convex set (crescent)
x2 = np.concatenate([np.cos(theta), -np.cos(theta[::-1])])
y2 = np.concatenate([np.sin(theta), 0.5 * np.sin(theta[::-1])])
ax[1].fill(x2, y2, alpha=0.5, color='orange', label='Non-Convex Set')
ax[1].plot([0.5, -0.5], [0.5, -0.5], 'r--')
ax[1].set_title("Non-Convex Set")
ax[1].legend()
ax[1].axis('equal')

plt.suptitle("Convex vs Non-Convex Sets")
plt.tight_layout()
plt.show()
```

```python
x = np.linspace(-2, 2, 100)
f_convex = x**2
f_nonconvex = np.sin(x)

plt.figure(figsize=(8, 4))
plt.plot(x, f_convex, label='Convex: $f(x) = x^2$')
plt.plot(x, f_nonconvex, label='Non-Convex: $f(x) = \sin(x)$')
plt.title("Convex vs Non-Convex Functions")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Simple gradient descent on f(x) = x^2
x_vals = []
f_vals = []
x = 2.0
lr = 0.1
for i in range(20):
    x_vals.append(x)
    f_vals.append(x**2)
    grad = 2 * x
    x -= lr * grad

plt.plot(x_vals, f_vals, marker='o')
plt.title("Gradient Descent on $f(x) = x^2$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
