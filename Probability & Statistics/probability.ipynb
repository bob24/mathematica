### Bayes' Theorem Explained Intuitively
#### Bayes' Theorem Formula:

$$ P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} $$

Where:
- **P(A|B)**: Posterior probability (probability of A given B has occurred)
- **P(B|A)**: Likelihood (probability of B given A is true)
- **P(A)**: Prior probability of A
- **P(B)**: Total probability of B
---

### üß† Real-World Example: Medical Diagnosis
Suppose 1% of people have a disease (**P(Disease) = 0.01**). A test correctly identifies 99% of people with the disease (**P(Positive|Disease) = 0.99**) but also has a 5% false positive rate (**P(Positive|No Disease) = 0.05**). <br>
Q) What is the probability a person actually has the disease if they test positive?

```python
### Step-by-Step Calculation
P_disease = 0.01
P_no_disease = 1 - P_disease
P_pos_given_disease = 0.99
P_pos_given_no_disease = 0.05

# Total probability of testing positive
P_positive = (P_pos_given_disease * P_disease) + (P_pos_given_no_disease * P_no_disease)

# Apply Bayes' Theorem
P_disease_given_positive = (P_pos_given_disease * P_disease) / P_positive

print(f"Probability of having disease if test is positive: {P_disease_given_positive:.4f}")
```

Even with a positive result, there's only ~17% chance the person has the disease!
### üîç Count-Based Reasoning (Intuitive)
Imagine 10,000 people:
- 100 have the disease (1%) ‚Üí 99 test positive (true positives)
- 9,900 don‚Äôt have the disease ‚Üí 495 test positive (false positives)

So:
- Total positive tests = 99 + 495 = 594
- Only 99 of those actually have the disease ‚Üí $$ \frac{99}{594} \approx 0.1667 $$
---

### üìò Conditional Probability

#### üîë Definition:
$$  P(A|B) = \frac{P(A \cap B)}{P(B)}  $$

- It represents the probability of A happening **given** that B has happened.
- Foundation of Bayes' theorem.

---


### üé≤ Example: Drawing Cards
Suppose we draw a card from a deck. Let:
- A = card is King
- B = card is a face card (Jack, Queen, King)

Then:
- $$ P(A) = \frac{4}{52} = 0.0769 $$
- $$ P(B) = \frac{12}{52} = 0.2308 $$
- $$ P(A \cap B) = \frac{4}{52} because all Kings are face cards $$
- $$ P(A|B) = \frac{4/52}{12/52} = \frac{1}{3} $$

#### ‚úÖ Meaning:
If you‚Äôre told the card is a face card, there's a **1 in 3** chance it‚Äôs a King.

---

### üß© Law of Total Probability

#### üîë Definition:
$$ P(B) = P(B|A_1)P(A_1) + P(B|A_2)P(A_2) + \ldots + P(B|A_n)P(A_n) $$

Where $$ A_1, A_2, \ldots, A_n $$ are **mutually exclusive and exhaustive** events.

---

### üéØ Example:
Let‚Äôs say a student can come from 3 departments: Math (30%), CS (50%), and Physics (20%).
Probability that each type of student passes a test:
- \( P(Pass|Math) = 0.9 \)
- \( P(Pass|CS) = 0.8 \)
- \( P(Pass|Physics) = 0.7 \)

What is the total probability that a **randomly selected student passes**?

```python
P_math = 0.3
P_cs = 0.5
P_phy = 0.2

P_pass_given_math = 0.9
P_pass_given_cs = 0.8
P_pass_given_phy = 0.7

P_pass = (P_pass_given_math * P_math) + \
         (P_pass_given_cs * P_cs) + \
         (P_pass_given_phy * P_phy)

print(f"Total probability that a student passes: {P_pass:.2f}")
```

### üîó Independence of Events

#### üîë Definition:
Two events A and B are independent if:
$$ P(A \cap B) = P(A) \cdot P(B) $$

Or equivalently,
$$ P(A|B) = P(A) \text{ and } P(B|A) = P(B) $$

---

### üß™ Example:
Let A = rolling a 3 on a die ‚Üí $$ P(A) = \frac{1}{6} $$
Let B = flipping heads on a coin ‚Üí $$ P(B) = \frac{1}{2} $$

Since these actions don‚Äôt affect each other:
- $$ P(A \cap B) = \frac{1}{6} \cdot \frac{1}{2} = \frac{1}{12} $$
- $$ P(A|B) = P(A) $$

#### ‚úÖ Meaning:
The occurrence of one does **not** change the likelihood of the other.

---

### üß† Bayesian Inference (ML Context)

Bayesian inference is the process of **updating our beliefs** about a hypothesis using new data.

#### üîÅ Formula:
$$ P(H|D) = \frac{P(D|H)P(H)}{P(D)} $$
- **H**: Hypothesis
- **D**: Observed Data

---

### üìà ML Example: Spam Detection

We have:
- P(Spam) = 0.4
- P(Word="free"|Spam) = 0.8
- P(Word="free"|Not Spam) = 0.1

What is the probability a message is spam if it contains the word "free"?

```python
P_spam = 0.4
P_not_spam = 0.6
P_free_given_spam = 0.8
P_free_given_not_spam = 0.1

P_free = (P_free_given_spam * P_spam) + (P_free_given_not_spam * P_not_spam)
P_spam_given_free = (P_free_given_spam * P_spam) / P_free

print(f"P(Spam|free): {P_spam_given_free:.2f}")
```

#### ‚úÖ Meaning:
Given the word ‚Äúfree‚Äù appears, there‚Äôs **84%** chance the message is spam.

```python
# Machine probabilities
P_A = 0.5
P_B = 0.3
P_C = 0.2

# Soda and straw probabilities for each machine
P_soda_A, P_straw_A = 0.8, 0.7
P_soda_B, P_straw_B = 0.5, 0.6
P_soda_C, P_straw_C = 0.2, 0.9

# Total probability of getting a soda with a straw
P_soda_straw = (P_A * P_soda_A * P_straw_A) + \
               (P_B * P_soda_B * P_straw_B) + \
               (P_C * P_soda_C * P_straw_C)

print(f'P_soda_straw: {P_soda_straw}')

# Bayes‚Äô Theorem: What is P(A | Soda+Straw)
P_A_given_soda_straw = (P_A * P_soda_A * P_straw_A) / P_soda_straw

print(f"Probability it came from Machine A: {P_A_given_soda_straw:.3f}")
```

Even though there are 3 machines, getting a soda with a straw makes it most likely to be from Machine A (‚âà61%).
