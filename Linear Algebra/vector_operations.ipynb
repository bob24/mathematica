```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Introduction to Vectors
##### 1) A vector is an ordered list of numbers, representing magnitude and direction.
##### 2) Example: A 2D vector v = [3, 4] represents a point in a plane with direction and magnitude.

## Vector Addition
##### Geometric Meaning: Adding vectors means placing the second vectorâ€™s tail at the head of the first vector.


```python
# Vector Addition
v1 = np.array([2, 3])
v2 = np.array([4, 1])
v_sum = v1 + v2
print(f'v1: {v1}')
print(f'v2: {v2}')
print(f'v_sum: {v_sum}')

plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1')
plt.quiver(v1[0], v1[1], v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', label='v2')
plt.quiver(0, 0, v_sum[0], v_sum[1], angles='xy', scale_units='xy', scale=1, color='g', label='v1 + v2')

plt.xlim(-1, 7)
plt.ylim(-1, 7)
plt.grid()
plt.legend()
plt.title("Vector Addition")
plt.show()
```

## Vector Subtraction
##### Geometric Meaning: Subtracting a vector means adding its negative (i.e reversing its direction).


```python
# Vector Subtraction
v_diff = v1 - v2
print(f'v1: {v1}')
print(f'v2: {v2}')
print(f'v_diff: {v_diff}')

plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1')
plt.quiver(0, 0, -v2[0], -v2[1], angles='xy', scale_units='xy', scale=1, color='b', label='-v2')
plt.quiver(0, 0, v_diff[0], v_diff[1], angles='xy', scale_units='xy', scale=1, color='g', label='v1 - v2')

plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.grid()
plt.legend()
plt.title("Vector Subtraction")
plt.show()
```

## Scalar Multiplication
##### Geometric Meaning: Scaling a vector changes its length while keeping its direction unchanged.


```python
# Scalar Multiplication
scalar = 2
v_scaled = scalar * v1
print(f'v1: {v1}')
print(f'v_scaled: {v_scaled}')

plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1')
plt.quiver(0, 0, v_scaled[0], v_scaled[1], angles='xy', scale_units='xy', scale=1, color='g', label='2 * v1')

plt.xlim(-1, 7)
plt.ylim(-1, 7)
plt.grid()
plt.legend()
plt.title("Scalar Multiplication")
plt.show()
```

##### 1) Formula: v1 â€¢ v2 = |v1| |v2| cos(Î¸)
##### 2) If dot product is 0, vectors are perpendicular.
##### 3) Example: v1 = [2,3], v2 = [4,1], v1 â€¢ v2 = (2 x 4) + (3 x 1) = 8 + 3 = 11.
##### 4) Geometric Meaning: The dot product measures similarity between two vectors and relates to the cosine of the angle between them.


```python
# Define vectors
v1 = np.array([2, 3])
v2 = np.array([4, 1])

# Compute dot product
dot_product = np.dot(v1, v2)
print(f'v1: {v1}, v2: {v2}')
print(f"Dot Product: {dot_product}")

# Compute magnitudes
v1_mag = np.linalg.norm(v1)
v2_mag = np.linalg.norm(v2)

# Compute angle between them using dot product formula
cos_theta = dot_product / (v1_mag * v2_mag)
theta = np.arccos(cos_theta)  # In radians
print(f"Magnitude of v1: {v1_mag:.2f}, Magnitude of v2: {v2_mag:.2f}")
print(f"Cos(Î¸) = {cos_theta:.3f}, Î¸ = {np.degrees(theta):.2f} degrees")
```

## Cross Product
##### Geometric Meaning: The cross product results in a vector perpendicular to both inputs and helps in 3D space calculations.


```python
# Cross Product
v3 = np.array([1, 0, 0])
v4 = np.array([0, 1, 0])
cross_product = np.cross(v3, v4)
print("Cross Product of v3 and v4:", cross_product)
```

## Norms
##### Geometric Meaning: Norms measure vector length.<br>
##### Euclidean is the straight-line distance, while Manhattan sums absolute differences.


```python
# Norms
euclidean_norm = np.linalg.norm(v1)
manhattan_norm = np.sum(np.abs(v1))
print(f'v1: {v1}')
print("Euclidean Norm of v1:", euclidean_norm)
print("Manhattan Norm of v1:", manhattan_norm)
```

## Projection of v1 onto v2

##### Geometric Meaning: Projection finds the component of one vector along another.


```python
print(f'v1: {v1}, v2: {v2}')

# Compute projection of v1 onto v2
proj_v1_on_v2 = (dot_product / np.dot(v2, v2)) * v2 #we are scaling by a space on to which we project then multiplying by space onto
# which we project [need to cross check if my understanding is correct or not]
print(f"proj_v1_on_v2: {proj_v1_on_v2}")

# Plot vectors
plt.figure(figsize=(6,6))
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', label='v2')
plt.quiver(0, 0, proj_v1_on_v2[0], proj_v1_on_v2[1], angles='xy', scale_units='xy', scale=1, color='g', label='Projection of v1 on v2')

# Draw dashed line from v1 to its projection on v2
plt.plot([v1[0], proj_v1_on_v2[0]], [v1[1], proj_v1_on_v2[1]], 'g--')

# Plot settings
plt.xlim(-1, 6)
plt.ylim(-1, 6)
plt.grid()
plt.legend()
plt.title("Dot Product & Projection")
plt.show()
```

## Unit Vectors & Basis Vectors
##### Geometric Meaning: Unit vectors define direction, and basis vectors are the building blocks of vector spaces in ML.


```python
print(f'v1: {v1}')
print(f'np.linalg.norm(v1): {np.linalg.norm(v1)}')

# Unit Vectors & Basis Vectors
unit_v1 = v1 / np.linalg.norm(v1)
print("Unit Vector of v1:", unit_v1)

# Standard basis vectors in 2D
e1 = np.array([1, 0])
e2 = np.array([0, 1])

plt.quiver(0, 0, e1[0], e1[1], angles='xy', scale_units='xy', scale=1, color='r', label='e1')
plt.quiver(0, 0, e2[0], e2[1], angles='xy', scale_units='xy', scale=1, color='b', label='e2')

plt.xlim(-1, 2)
plt.ylim(-1, 2)
plt.grid()
plt.legend()
plt.title("Unit & Basis Vectors")
plt.show()
```
